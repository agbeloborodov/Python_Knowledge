```python
'''Задача 44: В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?'''
```

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(n=20)
```

```python
''' 1) Решение по требованию условия задачи'''
```

```python
lst_zero = [0 for i in range(0,20)]
data1 = {'It is_human':lst_zero, 'It is_robot':lst_zero}
data1 = pd.DataFrame(data1, columns=['It is_human', 'It is_robot'])
data1['It is_human'].where(~(data.whoAmI == 'human'), other=1, inplace = True)
data1['It is_robot'].where(~(data.whoAmI == 'robot'), other=1, inplace = True)
data1.head(n=20)
```

```python
''' 2) Сравним результат с использованием get_dummies: '''
pd.get_dummies(data, prefix=['It is'])
```
